```python

def create_ground_truth_dist(timestamp,sniff):
  timestamp = timestamp.replace(tzinfo=None)
  
  mult = 0.6

  if sniff == 1:
      sniffer_pos = np.array([5,25])*mult
  if sniff == 2:
      sniffer_pos = np.array([20,7])*mult
  if sniff == 3:
      sniffer_pos = np.array([0,1])*mult
  
  # if the timestamp is btwn A and B the distance is C, kind of loke a TOA
  if (timestamp > pd.to_datetime('2022-02-28 13:22:00')) & (timestamp < pd.to_datetime('2022-02-28 13:22:21')):
    return (np.linalg.norm(sniffer_pos - np.array([4,4])*mult))
  if (timestamp > pd.to_datetime('2022-02-28 13:22:32')) & (timestamp < pd.to_datetime('2022-02-28 13:22:50')):
    return (np.linalg.norm(sniffer_pos - np.array([8,4])*mult))
  if (timestamp > pd.to_datetime('2022-02-28 13:23:02')) & (timestamp < pd.to_datetime('2022-02-28 13:23:20')):
    return (np.linalg.norm(sniffer_pos - np.array([12,4])*mult))
  if (timestamp > pd.to_datetime('2022-02-28 13:23:31')) & (timestamp < pd.to_datetime('2022-02-28 13:23:51')):
    return (np.linalg.norm(sniffer_pos - np.array([16,4])*mult))
  if (timestamp > pd.to_datetime('2022-02-28 13:24:04')) & (timestamp < pd.to_datetime('2022-02-28 13:24:20')):
    return (np.linalg.norm(sniffer_pos - np.array([10,8])*mult))
  if (timestamp > pd.to_datetime('2022-02-28 13:24:33')) & (timestamp < pd.to_datetime('2022-02-28 13:24:51')):
    return (np.linalg.norm(sniffer_pos - np.array([4,12])*mult))
  if (timestamp > pd.to_datetime('2022-02-28 13:24:56')) & (timestamp < pd.to_datetime('2022-02-28 13:25:16')):
    return (np.linalg.norm(sniffer_pos - np.array([8,12])*mult))
  if (timestamp > pd.to_datetime('2022-02-28 13:25:20')) & (timestamp < pd.to_datetime('2022-02-28 13:25:43')):
    return (np.linalg.norm(sniffer_pos - np.array([12,12])*mult))
  if (timestamp > pd.to_datetime('2022-02-28 13:26:00')) & (timestamp < pd.to_datetime('2022-02-28 13:26:18')):
    return (np.linalg.norm(sniffer_pos - np.array([16,12])*mult))
  if (timestamp > pd.to_datetime('2022-02-28 13:26:33')) & (timestamp < pd.to_datetime('2022-02-28 13:26:51')):
    return (np.linalg.norm(sniffer_pos - np.array([10,16])*mult))
  if (timestamp > pd.to_datetime('2022-02-28 13:27:04')) & (timestamp < pd.to_datetime('2022-02-28 13:27:22')):
    return (np.linalg.norm(sniffer_pos - np.array([4,20])*mult))
  if (timestamp > pd.to_datetime('2022-02-28 13:27:33')) & (timestamp < pd.to_datetime('2022-02-28 13:27:51')):
    return (np.linalg.norm(sniffer_pos - np.array([8,20])*mult))
  if (timestamp > pd.to_datetime('2022-02-28 13:28:03')) & (timestamp < pd.to_datetime('2022-02-28 13:28:21')):
    return (np.linalg.norm(sniffer_pos - np.array([12,20])*mult))
  if (timestamp > pd.to_datetime('2022-02-28 13:28:34')) & (timestamp < pd.to_datetime('2022-02-28 13:28:52')):
    return (np.linalg.norm(sniffer_pos - np.array([16,20])*mult))

def create_ground_truth(timestamp):
  timestamp = timestamp.replace(tzinfo=None)
  if (timestamp > pd.to_datetime('2022-02-28 13:22:00')) & (timestamp < pd.to_datetime('2022-02-28 13:22:21')):
    return 1
  if (timestamp > pd.to_datetime('2022-02-28 13:22:32')) & (timestamp < pd.to_datetime('2022-02-28 13:22:50')):
    return 2
  if (timestamp > pd.to_datetime('2022-02-28 13:23:02')) & (timestamp < pd.to_datetime('2022-02-28 13:23:20')):
    return 3
  if (timestamp > pd.to_datetime('2022-02-28 13:23:31')) & (timestamp < pd.to_datetime('2022-02-28 13:23:51')):
    return 4
  if (timestamp > pd.to_datetime('2022-02-28 13:24:04')) & (timestamp < pd.to_datetime('2022-02-28 13:24:20')):
    return 5
  if (timestamp > pd.to_datetime('2022-02-28 13:24:33')) & (timestamp < pd.to_datetime('2022-02-28 13:24:51')):
    return 6
  if (timestamp > pd.to_datetime('2022-02-28 13:24:56')) & (timestamp < pd.to_datetime('2022-02-28 13:25:16')):
    return 7
  if (timestamp > pd.to_datetime('2022-02-28 13:25:20')) & (timestamp < pd.to_datetime('2022-02-28 13:25:43')):
    return 8
  if (timestamp > pd.to_datetime('2022-02-28 13:26:00')) & (timestamp < pd.to_datetime('2022-02-28 13:26:18')):
    return 9
  if (timestamp > pd.to_datetime('2022-02-28 13:26:33')) & (timestamp < pd.to_datetime('2022-02-28 13:26:51')):
    return 10
  if (timestamp > pd.to_datetime('2022-02-28 13:27:04')) & (timestamp < pd.to_datetime('2022-02-28 13:27:22')):
    return 11
  if (timestamp > pd.to_datetime('2022-02-28 13:27:33')) & (timestamp < pd.to_datetime('2022-02-28 13:27:51')):
    return 12
  if (timestamp > pd.to_datetime('2022-02-28 13:28:03')) & (timestamp < pd.to_datetime('2022-02-28 13:28:21')):
    return 13
  if (timestamp > pd.to_datetime('2022-02-28 13:28:34')) & (timestamp < pd.to_datetime('2022-02-28 13:28:52')):
    return 14
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math

# PREPARE DATA TO BE ANALISE
fp_sniff_1 = pd.read_csv('fp_sniff_1.csv')
# pega todos os pacotes do dispositivo 1 (mac address) e dentro de  um length
fp_sniff_1 = fp_sniff_1.loc[(fp_sniff_1['wlan.sa'] == 'e0:dc:ff:32:e7:2f') & (fp_sniff_1['frame.len'] > 100)]
fp_sniff_1 = fp_sniff_1.drop(['frame.number','wlan.sa','frame.len'],axis=1)
# com base nas func anteriores calcula a distancia do dispositivo com base no timestamp dos pactoes
fp_sniff_1['gt_distance'] = fp_sniff_1['frame.time'].apply(lambda x : create_ground_truth_dist(pd.to_datetime(x),1))
# com base no timestamp sabe de cual punto viene -> loc 1,2,3,4 etc...
fp_sniff_1['gt_index'] = fp_sniff_1['frame.time'].apply(lambda x : create_ground_truth(pd.to_datetime(x))
)
fp_sniff_1 = fp_sniff_1[fp_sniff_1['gt_index'].notna()]
```

```python
# PROCESS THE DATA
from IPython.display import display
# obtem a media do signla strengh por location
#   mean da distancia e do dbm
print(fp_sniff_1.groupby(['gt_index']))
data = fp_sniff_1.groupby(['gt_index']).mean(numeric_only=True)
display(data)

plt.plot(data['gt_distance'],data['wlan_radio.signal_dbm'],'ob');
plt.xlabel('Distance [m]');
plt.ylabel('RSSI [dBm]');

# model stimation where the distance is based on the log function due to RSS
rssi = data['wlan_radio.signal_dbm']
distance = data['gt_distance']
log_d = data['gt_distance'].apply(np.log10)
z = np.polyfit(log_d, rssi, 1)
dd = np.linspace(0.5,20,40)
plt.plot(dd, z[1] + z[0]*np.log10(dd),'--r')

print('Propagation model: S = p0 + p1*log10(D)')
print('p0: %.2f' %z[1])
print('p1: %.2f' %z[0])

```

```python
import pandas as pd
import numpy as np

fp_sniff_1 = pd.read_csv('fp_sniff_1.csv')
fp_sniff_2 = pd.read_csv('fp_sniff_2.csv')
fp_sniff_3 = pd.read_csv('fp_sniff_3.csv')

# get the frames from each anchor node
#filter by mac: TAKE ONLY THE MAC INFOS OF THAT DEVICE
fp_sniff_1 = fp_sniff_1.loc[(fp_sniff_1['wlan.sa'] == 'e0:dc:ff:32:e7:2f') & (fp_sniff_1['frame.len'] > 100)]
fp_sniff_1 = fp_sniff_1.drop(['frame.number','wlan.sa','frame.len'],axis=1) #drop columns, protocol info not necessary
# rename some infos: sequence, time and dmb
fp_sniff_1 = fp_sniff_1.rename(columns = {'wlan.seq': 'sn_1', 'frame.time' : 'time_1', 'wlan_radio.signal_dbm' : 'rss_1'})
fp_sniff_1['time_1'] = pd.to_datetime(fp_sniff_1['time_1'])

fp_sniff_2 = fp_sniff_2.loc[(fp_sniff_2['wlan.sa'] == 'e0:dc:ff:32:e7:2f') & (fp_sniff_2['frame.len'] > 100)]
fp_sniff_2 = fp_sniff_2.drop(['frame.number','wlan.sa','frame.len'],axis=1)
fp_sniff_2 = fp_sniff_2.rename(columns = {'wlan.seq': 'sn_2', 'frame.time' : 'time_2', 'wlan_radio.signal_dbm' : 'rss_2'})
fp_sniff_2['time_2'] = pd.to_datetime(fp_sniff_2['time_2'])

fp_sniff_3 = fp_sniff_3.loc[(fp_sniff_3['wlan.sa'] == 'e0:dc:ff:32:e7:2f') & (fp_sniff_3['frame.len'] > 100)]
fp_sniff_3 = fp_sniff_3.drop(['frame.number','wlan.sa','frame.len'],axis=1)
fp_sniff_3 = fp_sniff_3.rename(columns = {'wlan.seq': 'sn_3', 'frame.time' : 'time_3', 'wlan_radio.signal_dbm' : 'rss_3'})
fp_sniff_3['time_3'] = pd.to_datetime(fp_sniff_3['time_3'])

# three different MEASSUREMENTS OF RSS from the same device
# join columns of RSS (1 from each acnhr) of the paches with the same sequence number
# vefore join row: seq_number, sn1/2/3, timestamp
# collect all the Rss of a seq packet in a row
# after join row: seq_number, sn1-timestamp, sn_2-timestamp, sn_3-timestamp
fp_join = fp_sniff_1.set_index('sn_1')
fp_join = fp_join.join(fp_sniff_2.set_index('sn_2'), how = 'outer')
fp_join = fp_join.join(fp_sniff_3.set_index('sn_3'))

#take the maximum timestamp: jus tto take a timestamp
fp_join['time'] = fp_join[['time_1','time_2','time_3']].max(axis=1)
fp_join['time'] = fp_join[['time_1','time_2','time_3']].max(axis=1)
fp_join = fp_join.drop(['time_1', 'time_2', 'time_3'], axis=1)
fp_join = fp_join.fillna(-100) # non read packets: put a very low number to act as far anchor
fp_join = fp_join.drop_duplicates()

#add ground truth - the location id column
fp_join['gt'] = fp_join['time'].apply(create_ground_truth)
fp_join = fp_join.dropna()
fp_join.to_csv('test.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, ConfusionMatrixDisplay

scaler = StandardScaler();

X = fp_join.drop(['time'],axis=1).dropna();
Y = X['gt'];
X = X.drop('gt',axis=1);

X_norm = scaler.fit_transform(X);

X_train, X_test, y_train, y_test = train_test_split(X_norm, Y, test_size=0.5)

ACCURACY = []
for k in range(1,5):
  knn = KNeighborsClassifier(n_neighbors=k, weights='distance')
  knn.fit(X_train, y_train)
  knn_predict = knn.predict(X_test)
  accuracy = accuracy_score(y_test, knn_predict)
  ACCURACY.append(accuracy)
  print(accuracy)

bestk = np.argmax(ACCURACY)+1
knn = KNeighborsClassifier(n_neighbors=bestk, weights='distance')
knn.fit(X_train, y_train)
knn_predict = knn.predict(X_test)
fig, ax = plt.subplots(figsize=(15, 15));
ConfusionMatrixDisplay.from_predictions(knn_predict, y_test, ax=ax, normalize='true');

fig = plt.figure();
plt.plot(range(1,5),ACCURACY);
plt.grid(visible=True);
plt.xlabel('K');
plt.ylabel('Accuracy');
```
